# Data Replacement, Mapping, Apply, and Assign in Pandas

## Learning Objectives

- Replace invalid or impossible values using replace()
- Handle multiple replacement conditions
- Use regex-based replacement
- Convert categorical values into numeric form
- Understand map() for Series transformation
- Use apply() for row/element-wise operations
- Create new columns using apply() and lambda
- Use assign() for functional-style DataFrame updates

```python
import numpy as np
import pandas as pd
```

## Loading Weather Dataset

- Dataset contains incorrect placeholder values
- Used to demonstrate data cleaning techniques

```python
df1 = pd.read_csv("weatherData.csv", encoding="latin-1")

df1
```

## Replacing Single Invalid Values

- replace() is used to convert invalid placeholders into NaN
- -99999 often represents missing or incorrect data

```python
new_df = df1.copy()

new_df.replace(-99999, np.nan, inplace=True)

new_df
```

## Replacing Multiple Invalid Values

- Multiple values can be replaced at once
- Useful for cleaning inconsistent datasets

```python
new_df = df1.copy()

new_df.replace([-99999, -88888], np.nan, inplace=True)

new_df
```

## Column-Specific Replacement Using Dictionary

- Different columns can have different replacement rules
- Helps in structured data cleaning

```python
new_df = df1.copy()

new_df.replace({
    "temperature": -99999,
    "windspeed": [-99999, -88888],
    "event": "0"
}, np.nan, inplace=True)

new_df
```

```python
df2 = pd.read_csv("weatherData2.csv", encoding="latin-1")

df2
```

## Mixed Value Replacement

- Can replace both numeric and text values together
- Useful for real-world messy datasets

```python
new_df = df2.copy()

new_df.replace({
    -99999: np.nan,
    "No Event": "Sunny"
}, inplace=True)

new_df
```

```python
df3 = pd.read_csv("weatherData3.csv", encoding="latin-1")

df3
```

## Regex-Based Replacement

- regex=True allows pattern-based replacement
- Useful for cleaning non-numeric characters from data

```python
new_df = df3.copy()

new_df.replace({
    "temperature": "[A-Za-z]",
    "windspeed": "[A-Za-z]"
}, "", regex=True, inplace=True)

new_df
```

## Replacing Multiple Categories with Numeric Values

- Converts categorical labels into numeric scores
- Useful for ML preprocessing

```python
df = pd.DataFrame({
    "Score": ["exceptional", "average", "good", "poor", "average", "exceptional"],
    "Student": ["Daniyal", "Umair", "Umar", "Raheem", "Hasnain", "Tariq"]
})

df.replace(
    ["poor", "average", "good", "exceptional"],
    [1, 2, 3, 4],
    inplace=True
)

df
```

## Mapping Values Using map()

- map() replaces values using a dictionary
- Works only on Series
- Commonly used for encoding categorical data

```python
df = pd.DataFrame({
    "Score": ["exceptional", "average", "good", "poor", "average", "exceptional"],
    "Student": ["Daniyal", "Umair", "Umar", "Raheem", "Hasnain", "Tariq"]
})

score_map = {"exceptional": 4, "average": 3, "good": 2, "poor": 1}

df["Score"] = df["Score"].map(score_map)

df
```

## Using apply() for Conditional Transformation

- apply() applies a function to each element
- Supports lambda functions
- Used for creating new derived columns

```python
df = pd.read_csv("employee_data.csv", encoding="latin-1")

df["Tax"] = df["Salary"].apply(lambda x: "20%" if x > 60000 else "10%")

df
```

## Using assign() to Create New Columns

- assign() adds new columns without modifying original DataFrame directly
- Returns a new DataFrame
- Useful for functional-style coding

```python
df = pd.read_csv("employee_data.csv", encoding="latin-1")

df = df.assign(New_Salary=df["Salary"] * 1.1)

df
```
