# Concatenation in Pandas

## Learning Objectives

- Combine multiple DataFrames using concat()
- Understand vertical (row-wise) concatenation
- Understand horizontal (column-wise) concatenation
- Use keys for hierarchical indexing
- Access individual concatenated sections
- Concatenate DataFrame with Series
- Understand index alignment in concatenation

```python
import pandas as pd
```

## Creating Sample DataFrames (Region-wise Data)

- Two separate DataFrames represent different regions
- Each contains customer information

```python
df1 = pd.DataFrame({
    "Customer Id": [1, 2],
    "Name": ["DANIYAL", "UMAIR"]
})

df2 = pd.DataFrame({
    "Customer Id": [3, 4],
    "Name": ["RAHEEM", "UMAR"]
})

df1
```

```python
df2
```

## Vertical Concatenation (Row-wise)

- axis=0 stacks DataFrames vertically
- This is the default behavior of concat()
- ignore_index=True resets index after merging

```python
df_concat = pd.concat([df1, df2], ignore_index=True)

df_concat
```

## Concatenation with Keys (Hierarchical Indexing)

- keys create a multi-level index
- Each DataFrame gets a label (region1, region2)
- Useful for identifying data sources after merging

```python
df_concat = pd.concat([df1, df2], keys=["region1", "region2"])

df_concat
```

## Accessing a Specific Region After Concatenation

- loc[] is used to access a specific key group
- Works only when keys are used

```python
df_concat.loc["region1"]
```

## Horizontal Concatenation (Column-wise)

- axis=1 combines DataFrames side by side
- Index alignment is important for correct merging

```python
temperature_df = pd.DataFrame({
    "city": ["Hyderabad", "Islamabad", "Karachi"],
    "temperature": [35, 25, 30]
}, index=[0, 1, 2])

windspeed_df = pd.DataFrame({
    "city": ["Islamabad", "Karachi", "Hyderabad"],
    "windspeed": [7, 12, 9]
}, index=[1, 2, 0])

temperature_df
```

```python
windspeed_df
```

## Concatenating DataFrames Horizontally

- axis=1 aligns data based on index values
- Columns are combined side by side

```python
df_concat = pd.concat([temperature_df, windspeed_df], axis=1)

df_concat
```

## Concatenating a Series with a DataFrame

- Series can also be combined with DataFrames
- It is treated as a new column
- Index alignment is applied automatically

```python
df = pd.DataFrame({
    "city": ["Islamabad", "Karachi", "Hyderabad"],
    "windspeed": [7, 12, 9]
})

s = pd.Series(["Rain", "Dry", "Humid"], name="Event")

df_concat = pd.concat([df, s], axis=1)

df_concat
```
