## What is Pandas

- Pandas is a Python library designed for data manipulation and data analysis
- It is used for cleaning, transforming, and structuring data
- It helps in finding patterns, trends, and insights from data
- Pandas is widely used in Data Science, Data Analysis, and Machine Learning

## What is a Series

- A Series is a one-dimensional labeled array
- It can store any data type such as integers, strings, floats, or boolean values
- Each element in a Series has a unique label called an index
- A Series is similar to a single column in a spreadsheet

## What is a DataFrame

- A DataFrame is a two-dimensional labeled data structure
- It is similar to an Excel spreadsheet, database table, or SQL table
- It consists of rows and columns
- Rows are identified by indices
- Columns are identified by column names
- A DataFrame is the most commonly used data structure in Pandas

## What is Data Munging or Data Wrangling

- Data Munging or Data Wrangling is the process of cleaning messy data
- It involves handling missing values, incorrect values, duplicate records, and formatting issues
- The goal is to convert raw data into a useful and meaningful format

```python
import pandas as pd
```

## Reading Data from CSV Files

- CSV stands for Comma Separated Values
- read_csv() is used to load CSV files into a DataFrame

```python
df = pd.read_csv("sales_data.csv", encoding="latin-1")
df
```

## Reading Data from Excel Files

```python
df = pd.read_excel("SampleSuperstore.xlsx")
df
```

## Reading Data from JSON Files

## Skipping Rows During File Import

- Some datasets contain unnecessary rows
- skiprows is used to ignore specific rows while reading a file

```python
df = pd.read_csv("stock_data1.csv", skiprows=1, encoding="latin-1")
df
```

## Setting a Custom Header Row

- Column names may not always appear in the first row
- header is used to specify which row contains the column names

```python
df = pd.read_csv("stock_data1.csv", header=1, encoding="latin-1")
df
```

## Reading Files Without Headers

- Some datasets do not contain column names
- header=None prevents Pandas from using the first row as column names

```python
df = pd.read_csv("stock_data2.csv", header=None, encoding="latin-1")
df
```

## Assigning Custom Column Names

- Column names can be assigned manually
- This improves readability and makes data easier to manage

```python
df = pd.read_csv(
    "stock_data2.csv",
    header=None,
    names=["ticker","eps","revenue","price","people"],
    encoding="latin-1"
)
df
```

## Reading a Limited Number of Rows

- nrows is used to load only a specific number of rows
- Useful when testing large datasets

```python
df = pd.read_csv("stock_data.csv", nrows=3, encoding="latin-1")
df
```

## Handling Missing Values

- Missing values may appear in different formats
- Pandas can automatically convert them into NaN values

```python
df = pd.read_csv(
    "stock_data.csv",
    na_values=["not available","n.a."],
    encoding="latin-1"
)
df
```

## Replacing Invalid Values with NaN

- Different columns may contain different invalid values
- Column-specific replacement rules can be defined using a dictionary

df = pd.read_csv(
    "stock_data.csv",
    na_values={
        "eps":["not available","n.a."],
        "price":["not available","n.a."],
        "people":["not available","n.a."],
        "revenue":["not available","n.a.",-1]
    },
    encoding="latin-1"
)
df

## Exporting Selected Columns to a CSV File

- Specific columns can be exported from a DataFrame
- index=False prevents row numbers from being written

```python
df = pd.read_csv("stock_data.csv", encoding="latin-1")

df.to_csv("stock_demo.csv", columns=["tickers","revenue","price"], index=False)
```

## Exporting Data Without Headers

- Header names can be excluded while exporting
- Useful for creating custom output files

```python
df = pd.read_csv("stock_data.csv", encoding="latin-1")

df.to_csv("stock_demo1.csv", header=False, index=False)
```

## Using Converters for Data Cleaning

- Converters allow custom processing during file import
- They help transform messy data into meaningful data

```python
def people_cell_converter(cell):
    if cell == "n.a.":
        return "sam walton"
    else:
        return cell

def eps_cell_converter(cell):
    if cell == "not available":
        return None
    else:
        return cell

df = pd.read_excel(
    "stock_data.xlsx",
    converters={
        "people": people_cell_converter,
        "eps": eps_cell_converter
    }
)

df
```

## Exporting Data to Excel Files

- DataFrames can be written to Excel files
- The starting row and column can be customized

```python
df = pd.read_excel("stock_data.xlsx")

df.to_excel("demo1.xlsx",startrow=1,startcol=2,index=False)
```

## Writing Multiple Worksheets in an Excel File

- ExcelWriter allows multiple DataFrames to be stored in one workbook
- Each DataFrame can be written to a separate worksheet

```python
df1 = pd.DataFrame( {
    "day" : ["1/1/2026","1/2/2026","1/3/2026","1/4/2026","1/5/2026","1/6/2026"],
    "temperature" : [32,35,28,24,32,31],
    "windspeed" : [6,7,2,7,4,2],
    "event" : ["Rain","Sunny","Snow","Snow","Rain","Sunny"]
})

df2 = pd.DataFrame( {
    "Name" : ["DANIYA ARAIN","UMAIR BHUTTO","RAHEEM KATIAR"],
    "Age" : [20,21,19],
    "City" : ["HYDERABAD","SUKKUR","HYDERABAD"]
})

with pd.ExcelWriter("combination.xlsx") as writer:
    df1.to_excel(writer,sheet_name="whether_data")
    df2.to_excel(writer,sheet_name="Student's data")
```
