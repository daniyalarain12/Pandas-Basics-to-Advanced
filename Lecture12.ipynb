# GroupBy and Aggregation in Pandas

## Learning Objectives

- Group data using single and multiple columns
- Apply aggregation functions like sum, mean, max
- Iterate through grouped data
- Access specific groups
- Use describe() on grouped data
- Apply multiple aggregations using agg()
- Rename aggregated columns

```python
import pandas as pd
```

```python
df = pd.DataFrame({
    "Name": ["DANIYAL", "UMAIR", "TARIQ", "RAHEEM", "UMAR"],
    "Age": [20, 24, 22, 24, 20],
    "Salary": [50000, 60000, 45000, 52000, 40000]
})

df
```

## Grouping Data by a Single Column

- groupby() is used to split data into groups
- Aggregation function sum() is applied to each group
- Here data is grouped by Age

```python
df.groupby("Age")["Salary"].sum()
```

## Grouping Data by Multiple Columns

- Data can be grouped using more than one column
- Each combination of Age and Name forms a group

```python
df.groupby(["Age", "Name"])["Salary"].sum()
```

```python
df = pd.read_csv("weather_by_cities.csv", encoding="latin-1")

df
```

## Finding Maximum Temperature by City

- groupby() splits data by city
- max() returns the highest temperature in each group

```python
df.groupby("city")["temperature"].max()
```

## Finding Average Wind Speed by City

- mean() calculates the average wind speed for each city

```python
df.groupby("city")["windspeed"].mean()
```

## Iterating Through Groups

- groupby() creates grouped objects
- Each group can be iterated using a loop
- Each iteration returns a city name and its corresponding DataFrame

```python
g = df.groupby("city")

for city, city_df in g:
    print(city)
    print(city_df)
```

## Accessing a Specific Group

- get_group() retrieves data for a specific group
- Useful when working with large grouped datasets

```python
g.get_group("mumbai")
```

## Descriptive Statistics for Each Group

- describe() provides summary statistics for each group
- Includes count, mean, min, max, and quartiles

```python
g.describe()
```

## Applying Multiple Aggregations on One Column

- agg() allows multiple aggregation functions
- Useful for detailed analysis of a single column

```python
df.groupby("city")["temperature"].agg(["mean", "max", "min"])
```

## Applying Multiple Aggregations with Custom Column Names

- Column names can be renamed during aggregation
- Makes output more readable and meaningful

```python
df.groupby("city")["temperature"].agg(avg_temperature="mean",max_temperature="max",min_temperature="min")
```

## Aggregation on Multiple Columns

- Different aggregation functions can be applied to different columns
- Each column can have its own function

```python
df.groupby("city").agg({
    "temperature": "mean",
    "windspeed": "max"
})
```

## Aggregation with Multiple Columns and Custom Names

- Output columns can be renamed while aggregating
- Tuple format is used: (column, function)

```python
df.groupby("city").agg(
    avg_temperature=("temperature", "mean"),
    max_windspeed=("windspeed", "max")
)
```
