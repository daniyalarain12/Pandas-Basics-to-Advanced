# Adding, Modifying, and Deleting Columns in Pandas

## Learning Objectives

- Add new columns to a DataFrame
- Insert columns at specific positions
- Modify individual cell values
- Update an entire column
- Delete single and multiple columns

```python
import pandas as pd
```

```python
df = pd.DataFrame({
    "Name": ["DANIYAL", "UMAR", "UMAIR", "RAHEEM", "HASNAIN", "QAYOOM", "NABEEL", "PREM"],
    "Age": [20, 24, 16, 56, 34, 46, 39, 28],
    "Salary": [100000, 40000, 85000, 44000, 90000, 34000, 45000, 78000],
    "Performance Score": [85, 90, 78, 92, 88, 95, 80, 89]
})

df
```

## Adding a New Column at the End

- A new column can be added by assigning values to a new column name
- The column is automatically added at the end of the DataFrame
- Here, Bonus is calculated as 10% of Salary

```python
df["Bonus"] = df["Salary"] * 0.1

df
```

## Inserting a New Column at a Specific Position

- insert() is used to place a column at a specific index
- The first argument specifies the position
- Existing columns shift to the right

```python
df.insert(0,"Employee Id",[10,20,30,40,50,60,70,80])

df
```

## Modifying a Single Cell Value

- loc is used to access a specific cell
- The value can be updated by assigning a new value

```python
df.loc[1, "Salary"] = 80000

df
```

## Updating an Entire Column

- A complete column can be updated using vectorized operations
- The operation is applied to all rows in the column
- Here, Salary is increased by 5%

```python
df["Salary"] = df["Salary"] * 1.05

df
```

## Deleting a Single Column

- drop() is used to remove columns from a DataFrame
- columns specifies which column should be deleted
- inplace=True updates the original DataFrame

```python
df.drop(columns="Performance Score",inplace=True)

df
```

## Deleting Multiple Columns

- Multiple columns can be removed by passing a list
- All specified columns are deleted together

```python
df.drop(columns=["Age","Salary"],inplace=True)

df
```
