# Working with Indexes in Pandas

## Learning Objectives

- Understand what indexes are
- View DataFrame indexes
- Set a custom index
- Reset indexes to default values

```python
import pandas as pd
```

```python
weather_data = {
    "day": ["1/1/2026", "1/2/2026", "1/3/2026", "1/4/2026", "1/5/2026", "1/6/2026"],
    "temperature": [32, 35, 28, 24, 32, 31],
    "windspeed": [6, 7, 2, 7, 4, 2],
    "event": ["Rain", "Sunny", "Snow", "Snow", "Rain", "Sunny"]
}

df = pd.DataFrame(weather_data)

df
```

## Displaying the Index Range

- Every DataFrame has an index
- By default, Pandas assigns integer indexes starting from 0
- The index attribute displays the complete index range

```python
df.index
```

## Setting a Custom Index

- Any column can be used as an index
- set_index() replaces the default index with values from the specified column
- inplace=True modifies the original DataFrame directly

```python
df.set_index("day", inplace=True)

df
```

## Resetting the Index

- reset_index() restores the default integer index

```python
df.reset_index(inplace=True)

df
```
