# Renaming, Ranking, and Reordering Data in Pandas

## Learning Objectives

- Rename column names
- Rename row labels
- Assign ranks to data
- Reorder columns
- Change the position of specific columns

```python
import pandas as pd
```

```python
df = pd.read_csv("employee_data.csv", encoding="latin-1")

df
```

## Renaming All Column Names

- The columns attribute can be used to replace all column names at once
- The number of new column names must match the number of existing columns

```python
df.columns = ["ID", "NAME", "AGE", "DEPARTMENT", "SALARY"]

df
```

## Renaming a Specific Column

- rename() is used to change selected column names
- Only the specified columns are renamed

```python
df.rename(columns={
    "SALARY" : "Income"
},inplace=True)

df
```

## Renaming a Specific Row Label

- Row labels can also be renamed
- rename() accepts a dictionary of old and new index labels

```python
df.rename(index={
    1 : "FIRST"
},inplace=True)

df
```

## Understanding rank()

- rank() assigns a position to each value within a column
- Ranking starts from 1
- Equal values receive the same rank
- It is commonly used in reporting and analytics

## Assigning Ranks in Ascending Order

- rank() assigns rankings based on column values
- Smaller values receive lower ranks
- By default, ranking is performed in ascending order

```python
df["Ranking"] = df["Income"].rank()

df
```

## Assigning Ranks in Descending Order

- ascending=False assigns the highest rank to the largest value
- Useful when ranking performance, salary, or scores

```python
df["Ranking"] = df["Income"].rank(ascending=False)

df
```

## Reordering Columns

- Columns can be rearranged by providing a new column order
- The DataFrame returns columns in the specified sequence

```python
df = df[["NAME","ID","AGE","DEPARTMENT","Income","Ranking"]]

df
```

## Moving a Column to the End

- A custom column order can be created dynamically
- List comprehension is used to exclude the target column
- The target column is then added at the end

```python
new_col_order = [col for col in df.columns if (col != "ID")] + ["ID"]
df = df[new_col_order]

df
```
