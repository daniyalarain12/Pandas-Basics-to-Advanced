# Handling Missing Values, Interpolation, and Data Cleaning in Pandas

## Learning Objectives

- Detect missing values using isnull()
- Count missing values per column
- Remove missing data using dropna()
- Fill missing values using fillna()
- Apply forward fill and backward fill
- Use interpolation methods for numerical data
- Handle time-based interpolation
- Remove rows based on conditions (how, thresh)
- Reindex DataFrames with date ranges
- Detect and remove duplicate rows

```python
import pandas as pd
```

## Creating a Sample DataFrame with Missing Values

- This dataset contains NaN and None values
- It is used to demonstrate missing data handling techniques

```python
df = pd.DataFrame({
    "Name" : ["DANIYAL","UMAR","UMAIR","RAHEEM","HASNAIN","QAYOOM","NABEEL","PREM"],
    "Age" : [20,None,16,56,34,46,39,28],
    "Salary": [100000,40000,85000,44000,90000,34000,45000,78000],
    "Performance Score" : [85,None,78,92,88,95,80,89]
})

df
```

## Detecting Missing Values

- isnull() returns True where values are missing
- sum() counts missing values per column

```python
df.isnull()
```

```python
df.isnull().sum()
```

## Removing Missing Values Using dropna()

- dropna() removes rows or columns containing missing values
- axis=0 removes rows
- axis=1 removes columns
- inplace=True modifies the original DataFrame

```python
new_df = df.copy()

new_df.dropna(inplace=True)

new_df
```

## Removing Columns with Missing Values

- axis=1 removes columns containing NaN values

```python
new_df = df.copy()

new_df.dropna(axis=1,inplace=True)

new_df
```

## Filling Missing Values Using fillna()

- fillna() replaces missing values with a given value
- Can be used with constants or computed values

```python
new_df = df.copy()

new_df.fillna(0, inplace=True)

new_df
```

## Filling Missing Values with Mean

- Missing values can be replaced with statistical measures
- Mean is commonly used for numerical data

```python
new_df = df.copy()

new_df["Age"].fillna(new_df["Age"].mean(),inplace=True)
new_df["Performance Score"].fillna(new_df["Performance Score"].mean(),inplace=True)

new_df
```

## Loading Time Series Data

- parse_dates converts string dates into datetime format
- Useful for time-based analysis

```python
df = pd.read_csv("weather_data.csv", parse_dates=["day"], encoding="latin-1")

df.set_index("day", inplace=True)

df
```

## Filling Missing Values with Dictionary

- Different columns can have different fill values
- fillna() supports dictionary-based replacement

```python
new_df = df.copy()

new_df.fillna({
    "temperature": 0,
    "windspeed": 0,
    "event": "No event"
}, inplace=True)

new_df
```

## Forward Fill (ffill)

- ffill propagates previous valid value forward
- Useful for time series continuity

```python
new_df = df.copy()

new_df.ffill(inplace=True)

new_df
```

## Forward Fill (ffill) with Limit

- ffill fills missing values using the previous valid value
- It moves top to bottom in the column
- limit parameter controls how many NaN values are filled consecutively
- If limit=1, only one missing value in a row will be filled

```python
new_df = df.copy()

new_df.ffill(limit=1, inplace=True)

new_df
```

## Backward Fill (bfill)

- bfill uses next valid value to fill missing data
- Works in reverse direction of ffill

```python
new_df = df.copy()

new_df.bfill(inplace=True)

new_df
```

## Backward Fill (bfill) with Limit

- bfill fills missing values using the next valid value
- It moves bottom to top in the column
- limit parameter controls how many consecutive NaN values are filled
- If limit=1, only one missing value in a sequence is filled

```python
new_df = df.copy()

new_df.bfill(limit=1, inplace=True)

new_df
```

## Interpolation (Linear Method)

- Interpolation estimates missing values mathematically
- Linear interpolation assumes straight-line change between values

```python
df = pd.DataFrame({
    "Time": [1, 2, 3, 4, 5],
    "Value": [10, None, 30, None, 50]
})

df.interpolate(method="linear", inplace=True)

df
```

```python
df = pd.read_csv("weather_data.csv",parse_dates=["day"],encoding="latin-1")

df.set_index("day",inplace=True)

df
```

## Interpolation (Default Method - Linear)

- interpolate() fills missing values using mathematical estimation
- Default method is linear interpolation
- It assumes values change in a straight line between points
- Works mainly on numerical data

```python
new_df = df.copy()

new_df.interpolate(inplace=True)

new_df
```

## Time-Based Interpolation

- method="time" considers time gaps between values
- Works only with datetime index

```python
new_df = df.copy()

new_df.interpolate(method="time",inplace=True)

new_df
```

## Dropping Rows Based on Conditions

- how="all" removes rows where all values are missing
- thresh keeps rows with minimum number of valid values

```python
new_df = df.copy()

new_df.dropna(how="all", inplace=True)

new_df
```

```python
new_df = df.copy()

new_df.dropna(thresh=1, inplace=True)

new_df
```

```python
new_df = df.copy()

new_df.dropna(thresh=2, inplace=True)

new_df
```

## Reindexing with Date Range

- reindex() aligns DataFrame with new index
- date_range creates continuous date sequence

```python
dt = pd.date_range("01-01-2017", "01-11-2017")
idx = pd.DatetimeIndex(dt)

df = df.reindex(idx)

df
```
