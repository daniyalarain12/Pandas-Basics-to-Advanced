# Data Selection with loc, iloc, at, and iat

## Learning Objectives

- Access rows using labels
- Access rows using positions
- Access individual cells
- Slice rows using labels and positions
- Select specific rows and columns
- Understand the difference between loc and iloc
- Understand the difference between at and iat

## Label-Based vs Position-Based Selection

- loc uses row and column labels
- iloc uses row and column positions
- at is a faster version of loc for accessing a single value
- iat is a faster version of iloc for accessing a single value

```python
import pandas as pd
```

```python
weather_data = {
    "day": ["1/1/2026", "1/2/2026", "1/3/2026", "1/4/2026", "1/5/2026", "1/6/2026"],
    "temperature": [32, 35, 28, 24, 32, 31],
    "windspeed": [6, 7, 2, 7, 4, 2],
    "event": ["Rain", "Sunny", "Snow", "Snow", "Rain", "Sunny"]
}

df = pd.DataFrame(weather_data)

df
```

## Setting a Custom Index

- The day column is used as the DataFrame index
- Label-based selection becomes easier with a meaningful index

```python
df.set_index("day", inplace=True)

df
```

## Accessing a Single Row Using loc

- loc is used for label-based selection
- Row labels are used instead of numeric positions

```python
df.loc["1/1/2026"]
```

## Accessing a Single Row Using iloc

- iloc is used for position-based selection
- Numeric indexes are used instead of labels

```python
df.iloc[0]
```

## Accessing a Single Cell Using loc

- Both row and column labels are specified
- loc returns the value stored at the specified location

```python
df.loc["1/1/2026", "event"]
```

## Accessing a Single Cell Using iloc

- Both row and column positions are specified
- iloc returns the value stored at the specified position

```python
df.iloc[0, 2]
```

## Slicing Rows Using loc

- loc performs label-based slicing
- The ending label is included in the result

```python
df.loc["1/1/2026":"1/4/2026"]
```

## Slicing Rows Using iloc

- iloc performs position-based slicing
- The ending position is excluded from the result

```python
df.iloc[0:4]
```

## Selecting Specific Rows and Columns Using loc

- Row labels and column names are used together
- Multiple columns can be selected by passing a list

```python
print(df.loc["1/1/2026":"1/4/2026",["temperature","windspeed"]])
```

## Selecting Specific Rows and Columns Using iloc

- Row and column positions are used together
- Position-based indexing is useful when labels are unknown

```python
df.iloc[0:4, 0:2]
```

## Fast Access to a Single Value Using at

- at is optimized for accessing a single value
- It uses row and column labels
- It is faster than loc when retrieving a single cell

```python
df.at["1/1/2026", "temperature"]
```

## Fast Access to a Single Value Using iat

- iat is optimized for accessing a single value
- It uses row and column positions
- It is faster than iloc when retrieving a single cell

```python
df.iat[0, 0]
```
