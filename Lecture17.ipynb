# Crosstab and Contingency Tables in Pandas

## Learning Objectives

- Understand contingency tables
- Use pd.crosstab() for categorical analysis
- Analyze relationships between categorical variables
- Add row/column totals using margins
- Compute percentage distributions using normalize
- Apply aggregation functions with values and aggfunc

```python
import pandas as pd
```

```python
df = pd.read_excel("survey.xls")

df
```

## Crosstab Between Nationality and Handedness

- crosstab() creates a frequency table
- It shows counts of combinations between two categorical columns

```python
pd.crosstab(df["Nationality"], df["Handedness"])
```

## Crosstab Between Sex and Handedness

- Shows distribution of handedness across genders

```python
pd.crosstab(df["Sex"], df["Handedness"])
```

## Adding Row and Column Totals

- margins=True adds totals for rows and columns
- Helps in overall distribution analysis

```python
pd.crosstab(df["Sex"], df["Handedness"], margins=True)
```

## Crosstab with Multiple Columns

- Multiple categorical variables can be analyzed together
- This creates a hierarchical column structure

```python
print(pd.crosstab(df["Sex"],[df["Handedness"], df["Nationality"]]))
```

## Crosstab with Multiple Row Variables

- Multiple row categories can be used together
- Useful for deeper segmentation analysis

```python
print(pd.crosstab( [df["Nationality"],df["Sex"]], df["Handedness"]))
```

## Crosstab with Percentage Distribution

- normalize="index" converts counts into percentages
- Each row sums to 100%
- Useful for comparative analysis

```python
pd.crosstab(df["Sex"], df["Handedness"], normalize="index")
```

## Crosstab with Aggregation Function

- values parameter allows numerical aggregation
- aggfunc applies statistical functions like mean
- Useful for combining categorical and numerical analysis

```python
print(pd.crosstab(df["Sex"],df["Handedness"],values=df["Age"],aggfunc="mean"))
```
