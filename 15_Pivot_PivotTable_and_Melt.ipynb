# Pivot, Pivot Table, and Melt in Pandas

## Learning Objectives

- Transform data using pivot (long → wide format)
- Use pivot_table for aggregation and summarization
- Apply grouping with pivot_table
- Use Grouper for time-based aggregation
- Reshape data using melt (wide → long format)
- Customize melted column names
- Understand structured data reshaping techniques

```python
import pandas as pd
```

```python
df = pd.read_csv("weather.csv")

df
```

## Pivot: Reshaping Data (Long → Wide Format)

- pivot() transforms long-form data into wide format
- index becomes rows
- columns become column headers
- values fill the table

```python
df.pivot(index="date", columns="city")
```

## Pivot with Specific Values

- values parameter selects which column to display
- Only selected metric is shown in pivot table

```python
df.pivot(index="date", columns="city", values="humidity")
```

## Pivot Table for Aggregation

- pivot_table() is used for summarization
- It automatically handles duplicate values using aggregation
- Default aggregation function is mean

```python
df.pivot_table(index="city", columns="date", aggfunc="max")
```

## Pivot Table with Margins

- margins=True adds row and column totals
- Provides overall aggregated values

```python
df.pivot_table(index="city", columns="date", margins=True)
```

## Time-Based Grouping Using Grouper

- Grouper is used for time-based aggregation
- freq="M" groups data by month
- Requires datetime format for proper grouping

```python
df = pd.read_csv("weather3.csv", parse_dates=["date"])

df
```

```python
df.pivot_table(index=pd.Grouper(freq="M",key="date"),columns="city")
```

## Reshaping Data Using Melt (Wide → Long Format)

- melt() converts wide data into long format
- Useful for normalization and analysis

```python
df = pd.read_csv("days_weather.csv")

df
```

## Basic Melt Operation

- id_vars keeps identifier columns fixed
- Other columns are unpivoted into rows

```python
df1 = pd.melt(df, id_vars="day")

df1
```

## Filtering Melted Data

- Melted data can be filtered like a normal DataFrame
- Useful for extracting specific city data

```python
df1[df1["variable"] == "chicago"]
```

## Melt with Custom Column Names

- var_name renames variable column
- value_name renames value column
- Improves readability

```python
df2 = pd.melt(df,id_vars="day",var_name="city",value_name="temperature")

df2
```

## Advanced Melt Example with Multiple ID Variables

- Multiple columns can be preserved using id_vars list
- Specific columns can be melted using value_vars

```python
df = pd.DataFrame({
    "country": ["USA", "USA", "India", "India"],
    "year": [2020, 2021, 2020, 2021],
    "sales": [100, 120, 90, 110],
    "profit": [20, 25, 18, 22],
    "tax": [2, 4, 1, 3]
})

df
```

## Melt with Multiple Value Variables

- Multiple metrics can be melted at once
- var_name stores metric names
- value_name stores actual values

```python
pd.melt(
    df,
    id_vars=["country", "year"],
    value_vars=["sales", "profit"],
    var_name="metrics",
    value_name="value"
)
```
