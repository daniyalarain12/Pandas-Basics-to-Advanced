# Sorting Data in Pandas

## Learning Objectives

- Sort data using a single column
- Sort data in ascending order
- Sort data in descending order
- Sort data using multiple columns
- Understand the ascending parameter

```python
import pandas as pd
```

```python
df = pd.DataFrame({
    "Name": ["DANIYAL", "UMAIR", "TARIQ"],
    "Age": [20, 22, 19],
    "Salary": [20000, 10000, 30000]
})

df
```

## Sorting Data Based on a Single Column

- sort_values() is used to sort data in a DataFrame
- A specific column is provided as the sorting key
- ascending=True sorts data in ascending order
- inplace=True updates the original DataFrame

```python
df.sort_values("Age", ascending=True, inplace=True)

df
```

## Sorting Data in Descending Order

- ascending=False sorts values from largest to smallest
- The sorting is performed on the specified column

```python
df.sort_values("Age", ascending=False, inplace=True)

df
```

## Creating a DataFrame for Multi-Column Sorting

- Multiple columns can be used as sorting keys
- If values in the first column are the same, Pandas uses the next column for sorting

```python
df2 = pd.DataFrame({
    "Name": ["DANIYAL", "UMAIR", "TARIQ"],
    "Age": [20, 20, 20],
    "Salary": [20000, 10000, 30000]
})

df2
```

## Sorting Data Based on Multiple Columns

- Multiple column names are passed as a list
- The ascending parameter can also accept a list
- Each True or False value corresponds to a specific column
- Pandas first sorts by Age and then by Salary

```python
df2.sort_values(
    ["Age", "Salary"],
    ascending=[True, True],
    inplace=True
)

df2
```
