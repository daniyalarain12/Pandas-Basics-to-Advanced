# Time Series Analysis in Pandas

## Learning Objectives

- Understand time series data structure
- Work with datetime index
- Slice data using date ranges
- Perform resampling (monthly, yearly, etc.)
- Generate custom date ranges
- Use business day frequency
- Apply forward filling for missing dates
- Shift and lag time series data
- Calculate returns and changes in stock prices

```python
import pandas as pd
```

## Loading Time Series Data (Apple Stock)

- Date column is converted into datetime format
- Date is set as index for time-based operations

```python
df = pd.read_csv("Apple.csv", parse_dates=["Date"], index_col="Date")

df.head()
```

## Accessing Data Using Date Index

- Individual date selection is possible using loc[]
- Partial string indexing allows filtering by year/month

```python
df.loc["2017-01-10"]
```

```python
df.loc["2017-01"]
```

## Monthly Average Closing Price

- Mean closing price is calculated for a specific month
- Useful for trend analysis

```python
df.loc["2017-01", "Close"].mean()
```

## Resampling Time Series Data

- resample() groups data based on time intervals
- "M" represents monthly frequency
- Useful for aggregating time-based data

```python
df["Close"].resample("M").mean()
```

## Generating Custom Date Range

- date_range() creates a sequence of dates
- freq="B" represents business days (Monday–Friday)

```python
df = pd.read_csv("Aapl_no_dates.csv")

df.head()
```

```python
rng = pd.date_range(start="6/1/2017", end="6/30/2017", freq="B")

df.set_index(rng, inplace=True)

df.head()
```

## Slicing Time Series Data

- Date ranges can be used for filtering rows
- Useful for analyzing specific periods

```python
df["2017-06-01":"2017-06-10"]
```

## Calculating Mean for a Date Range

- Mean is calculated for selected time window

```python
df["2017-06-01":"2017-06-10"].Close.mean()
```

## Handling Missing Dates with asfreq()

- asfreq() ensures continuous date frequency
- Missing dates can be filled using forward fill (pad)

```python
df.asfreq("D", method="pad")
```

## Generating Business Day Ranges

- freq="B" generates business days only
- Useful for financial datasets

```python
rng = pd.date_range(start="1/1/2026", periods=30, freq="B")

rng
```

## Custom Business Calendar (US Holidays Example)

- CustomBusinessDay excludes holidays
- Useful for real-world financial modeling

```python
from pandas.tseries.holiday import USFederalHolidayCalendar
from pandas.tseries.offsets import CustomBusinessDay

usb = CustomBusinessDay(calendar=USFederalHolidayCalendar())
```

```python
df = pd.read_csv("Aapl_no_date.csv")

rng = pd.date_range(start="7/1/2017", end="7/21/2017", freq=usb)

df.set_index(rng, inplace=True)

df.head()
```

## Shifting and Lagging Data

- shift() moves data forward or backward in time
- Used for comparison between time periods

```python
df = pd.read_csv("Fb.csv", parse_dates=["Date"], index_col="Date")

df
```

## Applying Shift Function

- shift(1) moves data forward
- shift(-1) moves data backward

```python
df.shift(1)
```

```python
df.shift(-1)
```

## Calculating Price Changes

- Previous day price is created using shift()
- Difference shows daily change in stock price

```python
df["Previous Day Price"] = df["Price"].shift(1)

df["1 day change"] = df["Price"] - df["Previous Day Price"]

df
```

## Calculating Percentage Returns

- Shows percentage change over 5 days
- Useful for financial analysis

```python
df["5 day return %"] = (df["Price"] - df["Price"].shift(5)) * 100 / df["Price"].shift(5)

df
```
