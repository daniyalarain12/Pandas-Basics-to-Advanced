# Merging DataFrames in Pandas

## Learning Objectives

- Combine multiple DataFrames
- Understand different types of joins
- Perform Inner, Outer, Left, Right, and Cross Joins
- Identify the source of merged data
- Handle common column names using suffixes

```python
import pandas as pd
```

## Creating Customer and Order DataFrames

- Two DataFrames are created for demonstrating merge operations
- Both DataFrames contain a common column named Customer Id

```python
df1 = pd.DataFrame({
    "Customer Id": [1, 2, 4],
    "Name": ["DANIYAL", "UMAIR", "TARIQ"]
})

df2 = pd.DataFrame({
    "Customer Id": [1, 2, 3],
    "Order Amount": [250, 450, 350]
})

df1
```

```python
df2
```

## Inner Join

- Returns only matching records from both DataFrames
- Rows without matching values are excluded
- inner is the most commonly used join type

```python
merged = pd.merge(df1,df2,on="Customer Id",how="inner")

merged
```

## Outer Join

- Returns all rows from both DataFrames
- Non-matching values are filled with NaN
- Combines matching and non-matching records

```python
merged = pd.merge(df1,df2,on="Customer Id",how="outer")

merged
```

## Using Indicator with Outer Join

- indicator=True adds a column named _merge
- It identifies the source of each row
- left_only means the row exists only in the left DataFrame
- right_only means the row exists only in the right DataFrame
- both means the row exists in both DataFrames

```python
merged = pd.merge(df1,df2,on="Customer Id",how="outer",indicator=True)

merged
```

## Left Join

- Returns all rows from the left DataFrame
- Matching rows are taken from the right DataFrame
- Non-matching values are filled with NaN

```python
merged = pd.merge(df1,df2,on="Customer Id",how="left")

merged
```

## Right Join

- Returns all rows from the right DataFrame
- Matching rows are taken from the left DataFrame
- Non-matching values are filled with NaN

```python
merged = pd.merge(df1,df2,on="Customer Id",how="right")

merged
```

## Cross Join

- Creates all possible combinations of rows
- Every row from the first DataFrame is matched with every row from the second DataFrame
- Also known as the Cartesian Product

```python
merged = pd.merge(df1,df2,how="cross")

merged
```

## Merging DataFrames with Common Columns

- Both DataFrames contain columns with the same names
- The city column is used as the merge key
- Inner Join is the default merge type

```python
df1 = pd.DataFrame({
    "city": ["Karachi", "Hyderabad", "Islamabad", "Quetta"],
    "temperature": [30, 37, 25, 15],
    "humidity": [65, 68, 71, 75]
})

df2 = pd.DataFrame({
    "city": ["Hyderabad", "Karachi", "Peshawar"],
    "temperature": [30, 37, 25],
    "humidity": [65, 68, 71]
})

df1
```

```python
df2
```

## Merging on a Common Column

- Pandas automatically performs an Inner Join when how is not specified
- Only matching city values are included

```python
merged = pd.merge(df1,df2,on="city")

merged
```

## Handling Duplicate Column Names Using Suffixes

- Both DataFrames contain temperature and humidity columns
- suffixes are added to distinguish columns with identical names
- This prevents column name conflicts after merging

```python
merged = pd.merge(df1,df2,on="city",suffixes=["_left","_right"])

merged
```
