# Data Selection, Filtering, and Aggregation in Pandas

## Learning Objectives

- Select single and multiple columns
- Slice rows from a DataFrame
- Filter data using single and multiple conditions
- Use the query() method for filtering
- Perform aggregation operations
- Generate descriptive statistics
- Retrieve DataFrame structure information

```python
import pandas as pd
```

```python
data = {
    "Name": ["DANIYAL", "UMAR", "UMAIR", "RAHEEM", "HASNAIN", "QAYOOM", "NABEEL", "PREM"],
    "Age": [20, 24, 16, 56, 34, 46, 39, 28],
    "Salary": [100000, 40000, 85000, 44000, 90000, 34000, 45000, 78000],
    "Performance Score": [85, 90, 78, 92, 88, 95, 80, 89]
}

df = pd.DataFrame(data)

df
```

## Selecting a Single Column

- A single column can be selected using square brackets
- Selecting a single column returns a Series

```python
df["Name"]
```

## Selecting Multiple Columns

- Multiple columns can be selected by passing a list of column names
- Selecting multiple columns returns a DataFrame

```python
subset = df[["Name", "Performance Score"]]

subset
```

## Slicing Rows

- Slicing is used to select a range of rows
- The starting index is included
- The ending index is excluded

```python
df[2:5]
```

## Filtering Rows Using a Single Condition

- Rows can be filtered using comparison operators
- Only rows satisfying the condition are returned

```python
high_salary = df[df["Salary"] > 50000]

high_salary
```

## Finding the Employee with the Maximum Salary

- max() returns the highest value in a column
- The matching row is filtered from the DataFrame

```python
max_sal = df[df["Salary"] == df["Salary"].max()]

max_sal
```

## Displaying Specific Columns for the Employee with the Maximum Salary

- Multiple columns can be selected after filtering
- This helps focus only on the required information

```python
df[["Name", "Salary"]][df["Salary"] == df["Salary"].max()]
```

## Filtering Rows Using Multiple Conditions with AND

- The & operator represents the AND condition
- Parentheses are required around each condition
- Both conditions must be true

```python
filtered = df[ (df["Salary"] > 50000) & (df["Age"] > 25) ]

filtered
```

## Filtering Rows Using Multiple Conditions with OR

- The | operator represents the OR condition
- At least one condition must be true

```python
filtered2 = df[ (df["Age"] > 25) | (df["Performance Score"] > 90) ]

filtered2
```

## What is the query() Method

- query() is used to filter rows using string expressions
- It improves readability for complex conditions
- Column names can be used directly in expressions
- Backticks are used for column names containing spaces
- The @ symbol is used to reference Python variables

## Filtering Rows Using query()

- Conditions are written as strings
- The result is similar to Boolean indexing

```python
high_salary = df.query("Salary > 50000")

high_salary
```

## Finding the Employee with the Maximum Salary Using query()

- Aggregate functions can be used inside query()
- The matching row is returned

```python
max_sal = df.query("Salary == Salary.max()")

max_sal
```

## Displaying Specific Columns Using query()

- Filtering and column selection can be combined

```python
df.query("Salary == Salary.max()")[["Name", "Salary"]]
```

## Filtering Rows Using Multiple Conditions with query()

- and is used for logical AND
- Variables can be referenced using the @ symbol

```python
sal = 50000

filtered = df.query("Salary > @sal and Age > 25")

filtered
```

## Filtering Rows Using OR with query()

- or is used for logical OR
- Backticks are required for column names containing spaces

```python
filtered2 = df.query("Age > 25 or `Performance Score` > 90")

filtered2
```

## Aggregation Functions

- Aggregation functions summarize data
- They are commonly used for analysis and reporting

## Finding the Maximum Salary

- max() returns the largest value in a column

```python
df["Salary"].max()
```

## Finding the Minimum Salary

- min() returns the smallest value in a column

```python
df["Salary"].min()
```

## Calculating the Average Salary

- mean() returns the arithmetic average of values

```python
df["Salary"].mean()
```

## Calculating the Total Salary

- sum() returns the total of all values in a column

```python
df["Salary"].sum()
```

```python
df
```

## Generating Descriptive Statistics

- describe() generates summary statistics for numerical columns
- It displays count, mean, standard deviation, minimum, maximum, and quartiles

```python
df.describe()
```

## Displaying the Shape of a DataFrame

- shape returns the number of rows and columns
- The result is returned as a tuple
- The first value represents rows
- The second value represents columns

```python
df.shape
```

## Extracting the Number of Rows and Columns

- The shape tuple can be unpacked into separate variables
- This provides individual access to rows and columns

```python
rows, columns = df.shape

rows, columns
```

## Displaying Column Names

- columns returns all column names in the DataFrame
- It helps inspect the structure of the dataset

```python
df.columns
```
